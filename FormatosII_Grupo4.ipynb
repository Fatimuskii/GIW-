### Práctica de Formatos de datos: XML

El objetivo de esta práctica es crear una aplicación que gestionará información sobre los parques y jardines públicos de Madrid. Para ello se van a utilizar dos APIS diferentes ofrecidas por el portal de datos abiertos del ayuntamiento de Madrid.

En primer lugar se va a usar el catálogo "Principales parques y jardines municipales". Este catálogo ofrece una relación de los parques y zonas verdes cuya gestión corresponde al Ayuntamiento de Madrid, con datos de dirección y de localización para georeferenciarlos. El siguiente trozo de código permite recuperar el archivo xml en cuestion, y lo almacena en disco.

```python
import urllib.request
x = urllib.request.urlopen('https://datos.madrid.es/egob/catalogo/200761-0-parques-jardines.xml')
a=x.read()
a=a.decode("utf-8")
prueba=open("catalogo.xml","w")
prueba.write(str(a))
prueba.close()
```

```python
def mostrarParques():
    from xml.etree import ElementTree
    f=open("catalogo.xml","rt")
    arbol=ElementTree.parse(f)
    raiz= arbol.getroot()
    
    print("      Lista de Principales parques y Jardines municipales:"+ '\n')
   
    i=1
    for nodo in raiz.iter("atributo"):    
        for a in nodo.attrib:
            elem=nodo.attrib[a]
            if elem =="NOMBRE":
                print(i,nodo.text)        
                i+=1
        
    return i,raiz
    
```

```python
def mostrarDatos(num,raiz):

    i=0
    for nodo in raiz.iter("atributos"):
        if(i==num):
            dir="Direccion: \n"
            for nodoHijo in nodo.iter("atributo"):                
                for a in nodoHijo.attrib:
                    elem=nodoHijo.attrib[a]    
                    if "NOMBRE"== elem:
                        print("Nombre del parque:"+"\n   ",nodoHijo.text,"\n")        
                    elif "DESCRIPCION-ENTIDAD" in elem:
                        print("Descripción del parque: \n   ",nodoHijo.text,"\n")
                    elif "DESCRIPCION" in elem:
                        print("Información conservación: \n   ",nodoHijo.text,"\n")
                    elif "HORARIO" in elem:
                        print("Horario:\n"  ,nodoHijo.text,'\n')
                    elif "EQUIPAMIENTO" in elem:
                        print("Equipamiento:\n", nodoHijo.text, '\n')
                    elif "TRANSPORTE" in elem:
                        print("Transporte:\n", nodoHijo.text, '\n')
                    elif "NOMBRE-VIA" == elem:
                         dir+=nodoHijo.text+", "
                    elif "NUM" ==elem:
                        dir+=nodoHijo.text+". "
                    elif "PUERTA" ==elem:
                        dir+=nodoHijo.text+" "
                    elif "LOCALIDAD" == elem:
                        dir+=nodoHijo.text+". "
                    elif "PROVINCIA" ==elem:
                        dir+=nodoHijo.text+". "
                    elif "CODIGO-POSTAL"==elem:
                        dir+=nodoHijo.text+" "
                        print(dir, '.','\n')
                    elif "ACCESIBILIDAD"==elem:
                        print("Accesibilidad: " + nodoHijo.text,'\n')
                        
        i+=1 
```

```python
def pedirNumero(num, raiz):
    while True:
        opcion=(input("Indique el número del parque que desee consultar del 1-"+str(num-1)+"(0 PARA SALIR):\n"))
        
        if(int(opcion) ==0):
            break
        if opcion !="":
            mostrarDatos(int(opcion)-1,raiz)
```

```python
numMax,raiz= mostrarParques()
pedirNumero(numMax,raiz)
```

```python
#Funcion que busca los datos con los filtros dados por el usuario.

def buscarPor(accesibilidad, nombre, barrio, distrito,arbol):
  
    listaCriteria=[]
   
    if(accesibilidad!="NINGUNO"):
        listaCriteria.append(accesibilidad)
    if(nombre!="NINGUNO"):
        listaCriteria.append(nombre)
    if(barrio!="NINGUNO"):
        listaCriteria.append(barrio)
    if(distrito!="NINGUNO"):
        listaCriteria.append(distrito)
    
    neigh=""
    district=""
    listaValidos=[]

    i=0
    for nodo in arbol.findall("./contenido/atributos"):
        cont=0
        
        name = nodo.find("*[@nombre='NOMBRE']").text
        access=nodo.find("*[@nombre='ACCESIBILIDAD']").text
        
        if(nodo.find("atributo[@nombre='LOCALIZACION']/atributo[@nombre='BARRIO']")!=None):
            neigh=nodo.find("atributo[@nombre='LOCALIZACION']/atributo[@nombre='BARRIO']").text
            
        if(nodo.find("atributo[@nombre='LOCALIZACION']/atributo[@nombre='DISTRITO']")!=None):
            district=nodo.find("atributo[@nombre='LOCALIZACION']/atributo[@nombre='DISTRITO']").text
        
        if(name in listaCriteria):
            cont+=1
        if(access in listaCriteria):
            cont+=1
        if(neigh in listaCriteria):
            cont+=1
        if(district in listaCriteria):
            cont+=1
    
        if (cont == len(listaCriteria)):
            listaValidos.append(i)
            
        i+=1

    return listaValidos
```

```python
def imprimirResultados(listaValidos,raiz):
    if (len(listaValidos) == 0):
        print("NO HAY RESULTADOS", '\n')
        
    else:
        print("RESULTADOS DE SU BUSQUEDA - Numero de resultados:", str(len(listaValidos)),'\n')
        for i in listaValidos:
            mostrarDatos(i, raiz)
    
```

```python
def menu():
    
    from xml.etree import ElementTree
    f=open("catalogo.xml","rt")
    arbol=ElementTree.parse(f)
    raiz= arbol.getroot()
    
    accesibilidad=""
    nombre=""
    barrio=""
    distrito=""
    resultado=[]
    
    while True:
        print("MENU DE BUSQUEDA:\n")
        print("1. Seleccionar accesibilidad:", "Tipo " + accesibilidad, '\n')
        print("2. Seleccionar nombre:", nombre, '\n')
        print("3. Seleccionar barrio:", barrio, '\n')
        print("4. Añadir distrito:", distrito, '\n')
        print("5. Buscar:",'\n')
        print("0. Salir:", '\n')
        
        opcion=input("Opción:")
        if(opcion!=""):
            opcion=int(opcion)
            if(opcion==0):
                break
            elif (opcion==1):
                accesibilidad=str(mostrarAccesibilidad(raiz))
            elif (opcion==2):
                nombre=mostrarNombres(raiz)
            elif(opcion==3):
                barrio=mostrarBarrios(raiz)
            elif(opcion ==4):
                distrito=mostrarDistritos(raiz)
            elif (opcion ==5):
                if(accesibilidad ==""):
                    accesibilidad="NINGUNO"
                if(nombre ==""):
                    nombre="NINGUNO"
                if(barrio ==""):
                    barrio="NINGUNO"
                if(distrito==""):
                    distrito="NINGUNO"
            
                resultado = buscarPor(accesibilidad, nombre, barrio,distrito,arbol)
                imprimirResultados(resultado,raiz)
                accesibilidad=""
                nombre=""
                barrio=""
                distrito=""
            
            else:
                print("Opcion no válida. ")
                continue
        print("\n")         
        
```
