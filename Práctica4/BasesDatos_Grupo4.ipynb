```python
def crearTablas(conn):
    
    cur=conn.cursor()

    cur.execute("DROP TABLE IF EXISTS Marcas")
    cur.execute("DROP TABLE IF EXISTS Coches")
    cur.execute("DROP TABLE IF EXISTS Marco")
    cur.execute("DROP TABLE IF EXISTS Concesionarios")
    cur.execute("DROP TABLE IF EXISTS Distribucion")
    cur.execute("DROP TABLE IF EXISTS Clientes")
    cur.execute("DROP TABLE IF EXISTS Ventas")
    
    conn.commit()
    
    #Marca
    cur.execute("CREATE TABLE Marcas(cifm INTEGER(4) PRIMARY KEY, Nombre TEXT, Ciudad TEXT)")
     #Coches
    cur.execute("CREATE TABLE Coches(codCoche INTEGER(4) PRIMARY KEY, nombre TEXT, modelo TEXT)")
    #Marco
    cur.execute("CREATE TABLE Marco(cifm INTEGER(4) REFERENCES Marcas(cifm),"
                " codCoche INTEGER(4) REFERENCES Coches(codCoche), PRIMARY KEY (cifm,codCoche))")
    #Concesionarios
    cur.execute("CREATE TABLE Concesionarios(cifc INTEGER(4) PRIMARY KEY, nombre TEXT, ciudad TEXT)")
    #Distribución
    cur.execute("CREATE TABLE Distribucion(cifc INTEGER(4) REFERENCES Concesionarios(cifc),"
                " codCoche INTEGER(4) REFERENCES Coches(codCoche),"
                " cantidad INTEGER, PRIMARY KEY(cifc, codCoche))")
    #Clientes
    cur.execute("CREATE TABLE Clientes(dni INTEGER(4) PRIMARY KEY, nombre TEXT, apellidos TEXT, ciudad TEXT)")
    #Ventas
    cur.execute("CREATE TABLE Ventas(cifc INTEGER(4) REFERENCES Concesionarios(cifc),"
                " dni INTEGER(4) REFERENCES Clientes(dni),"
                " codCoche INTEGER(4) REFERENCES Coches(codCoche), color TEXT,"
                " PRIMARY KEY (cifc, dni, codCoche))")
    
    cur.close()
    conn.commit()
```

```python
import sqlite3
conn=sqlite3.connect("vehiculos.sqlite3")
cur=conn.cursor()
```

```python
#1.Obtener el nombre y el apellido de los clientes que han comprado como mínimo un coche de 
#color ‘verde’ o un coche de color ‘azul’
cur.execute("SELECT DISTINCT c.dni, c.nombre, c.apellidos "
            "FROM Clientes c JOIN Ventas v USING(dni) "
            "WHERE v.color = 'verde' OR v.color = 'azul'")

consulta=cur.fetchall()
if(not consulta):
    print ("No existen resultados para esa consulta.")
else:
    for row in consulta:
        print("Nombre y apellidos: "+ row[1]+ " "+row[2])
```

```python
#2.Obtener el nombre de los concesionarios que sólo han vendido coches al cliente con dni igual a 0001
cur.execute("SELECT c.nombre, COUNT(v.dni) AS count "
            "FROM Concesionarios c JOIN Ventas v USING(cifc) "
            "WHERE v.dni=0001 "
            "GROUP BY cifc HAVING count = (SELECT COUNT(dni) "
                                          "FROM Ventas ve "
                                          "WHERE ve.cifc = v.cifc "
                                          "GROUP BY cifc)")

consulta=cur.fetchall()
if(not consulta):
    print ("No existen resultados para esa consulta.")
else:
    for row in consulta:
        print("Concesionario: "+ row[0])
```

```python
#3.Obtener los codcoche y la media de los valores del atributo cantidad para cada codcoche 
#que aparece en la relación distribucion para los casos en que la media sea menor que 15
cur.execute("SELECT DISTINCT codCoche, AVG(cantidad) AS mediaCantidad "
            "FROM Distribucion "
            "GROUP BY codCoche HAVING mediaCantidad < 15")

consulta=cur.fetchall()
if(not consulta):
    print ("No existen resultados para esa consulta.")
else:
    for row in consulta:
        print("codCoche: "+ str(row[0])+ "\tmedia: "+ str(row[1]))
```

```python
#4.Obtener para cada valor de dni de los clientes el número de veces que aparece en la relación ventas.
cur.execute("SELECT c.dni, COUNT(v.dni) "
            "FROM Clientes c LEFT JOIN Ventas v USING(dni) "
            "GROUP BY c.dni")

consulta=cur.fetchall()
if(not consulta):
    print ("No existen resultados para esa consulta.")
else:
    for row in consulta:
        print("DNI: "+ str(row[0]) + " " + "Número de veces: "+ str(row[1]))
```

```python
#5.Obtener el dni de los clientes de nombre ‘pepe’ o de nombre ‘luis’ (no se admiten valores repetidos)
cur.execute("SELECT DISTINCT dni "
            "FROM Clientes "
            "WHERE LOWER(nombre)='pepe' OR LOWER(nombre)='luis'")

consulta=cur.fetchall()
if(not consulta):
    print ("No existen resultados para esa consulta.")
else:
    for row in consulta:
        print("DNI: "+ str(row[0]))
```

```python
#6.Obtener el cifc de todos los concesionarios cuyo atributo cantidad en la relación distribucion 
#no está comprendido entre 10 y 18 ambos inclusive (es decir, los valores menores que 10 y mayores que 18)
cur.execute("SELECT cifc, SUM(cantidad) AS sumaCantidad "
            "FROM Distribucion "
            "GROUP BY cifc "
            "HAVING sumaCantidad NOT BETWEEN 10 AND 18")

consulta=cur.fetchall()
if(not consulta):
    print ("No existen resultados para esa consulta.")
else:
    for row in consulta:
        print("Concesionario: "+ str(row[0]))
```
